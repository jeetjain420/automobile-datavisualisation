```python
def read_auto_data(fileName = "automobile"):
    'Function to load the auto price data set from a .csv file' 
    import pandas as pd
    import numpy as np

    ## Read the .csv file with the pandas read_csv method
    auto_prices = pd.read_csv(r"C:\Users\acer pc\OneDrive\Desktop\visualisation\automobile.csv")
    
    ## Remove rows with missing values, accounting for mising values coded as '?'
    cols = ['price', 'bore', 'stroke', 'horsepower', 'peak-rpm']
    for column in cols:
        auto_prices.loc[auto_prices[column] == '?', column] = np.nan
    auto_prices.dropna(axis = 0, inplace = True)

    ## Convert some columns to numeric values
    for column in cols:
        auto_prices[column] = pd.to_numeric(auto_prices[column])
#    auto_prices[cols] = auto_prices[cols].as_type(int64)
        
    return auto_prices
auto_prices = read_auto_data()
```

```python
fig = plt.figure(figsize = (6,6))

ax = fig.gca()
auto_prices['engine-size'].plot.hist(ax=ax)
ax.set_title('Histogram of auto engine size')
ax.set_xlabel('Engine Size')
ax.set_ylabel('No. of autos')
```

```python
fig = plt.figure(figsize = (8,6))
ax = fig.gca()
auto_prices[['engine-size', 'fuel-type']].boxplot(by = 'fuel-type', ax = ax)
ax.set_title('Box plot')
ax.set_xlabel('Fuel Type')
ax.set_ylabel('Engine Size')
```

After examining the plot we can see that the engine size of diesel cars is slightly larger than the gas cars but there are some outliers present in the gas cars which have a much higher size compared to the maximum size of the diesel cars

```python
import seaborn as sns
sns.set_style("whitegrid")
sns.kdeplot(auto_prices['engine-size'])
```

```python
fig = plt.figure(figsize=(6,6)) # define plot area
ax = fig.gca() # define axis 
sns.set_style("whitegrid")
sns.kdeplot(auto_prices['engine-size'], ax = ax)
ax.set_title('KDE plot of auto engine size') # Give the plot a main title
ax.set_xlabel('Engine Size') # Set text for the x axis
ax.set_ylabel('Density')# Set text for y axis
```

```python
fig = plt.figure(figsize = (7,7))
ax = fig.gca()
sns.set_style("whitegrid")
sns.kdeplot(auto_prices[['engine-size', 'price']], ax = ax, cmap ="Blues")
ax.set_title('KDE Plot')
ax.set_xlabel('Engine Size')
ax.set_ylabel('Price')
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.gca()
sns.set_style("whitegrid")
sns.violinplot(x = 'fuel-type', y = 'engine-size', data = auto_prices, ax=ax)
ax.set_title('Violin plots of auto engine size by fuel type')
ax.set_xlabel('Fuel Type')
ax.set_ylabel('Engine Size')
```
